### List vs Numpy array 

```python
py_list = [1,2,3]
print("Python list Multiplication: ", py_list * 2)  # length wise multiplication i,e repeying same list n times 

np_array = np.array([1,2,3])
print("Numpy array Multiplication: ",np_array * 2)    #elements wiise multiplication
```

### Creating Array from the Scratch

```python
zeros = np.zeros((3,4))
print(zeros) 

ones = np.ones((2,3))
print(ones)

constants = np.full((2,2) , 7)    #to create array with k in it 
print(constants)

rand = np.random.random((3,3))
print(rand)

sequence = np.arange(0,10,2)   # start , end , gap
print(sequence)
```

### Vector Matrix and Tensor

```python
vector = np.array([1,2,3,4])     #1-d array
print("Vector\n" , vector)

matrix = np.array([[1,2,3] , [4,5,6]])   #2-d array
print("Matrix\n" ,matrix)

tensor = np.array([[[1,2],[3,4]],[[5,6],[7,8]],[[8,9],[10,11]]])  #3-d array and more dimensionals all are tensors
print("Tensor\n" ,tensor)
```

### Array Properties >>> 
- no issue if u keep data of differ datatype 
- but to optimize processing its prefer to have all emelemts with same datatypes 

```python
arr = np.array([[1,2,3] , [4,5,6]])
print("Shape: ", arr.shape)    #m x n
print("Dimension: ", arr.ndim)   # 1d 2d or 3d and more
print("Size: ", arr.size)   # Total elements here 
print("Datatype: ",arr.dtype)
```

```python
arr = np.array([[1,2,3] , [4,5,"om"]]) 
print("Datatype: ",arr.dtype)

arr = np.array([[1,2,3] , [4,5,3.8]])
print("Datatype: ",arr.dtype)
```

### Array reshaping 

```python
arr = np.arange(12)
print("Original Array: ",arr)

reshaped = arr.reshape((3,4))
print("Reshaped Array: " , reshaped)

# reshaped = arr.reshape((3,3))     #This gives error as total elements should be matched before and after 
# print("Reshaped Array: " , reshaped)

flattened = reshaped.flatten()
print("Flatten Array: " , flattened)

#ravel returns view instead of the copy 
reveled = reshaped.ravel()
print("Raveled Array: ", reveled)

# Transpose of the matrixes 
print("Array: " , reshaped)
transpose = reshaped.transpose()
print("Transposed Array: " , transpose)
```
