## Advance Operation with Business Example

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Data Structure : [restaurent_id , 2021 , 2022 , 2023, 2024]
sales_data = np.array([
    [1,150000,180000,220000,250000],   #paradise Biryani
    [2,120000,140000,160000,190000],   #Beijing Bites
    [3,200000,230000,260000,300000],   #Pizza hub
    [4,180000,210000,240000,270000],   #Burger Point
    [5,160000,185000,205000,230000],   #Chai Point
])
```

```python
#First three restaurent data 
length = len(sales_data)
# print(length)

# for i in range(3):
#     print(f"Restaurent number {i+1}: " , sales_data[i])

#Numpy ka tarika
indexes = [0,1,2]
first_three = sales_data[indexes]
print(first_three)

first_three = sales_data[0:3]  #infact [:3]  also works 
print(first_three)
```

```python
#sum of the all the sales 
# colwise = np.sum(sales_data , axis = 0)
# print(colwise)                                                                                      
# total = np.sum(colwise) - colwise[0]
# print(total)

yearly_total = np.sum(sales_data[ : , 1:] , axis = 0)
print(yearly_total)
```

```python
#Minimum sales per the restaurent 
#To get across all the years and per the restaurent 
min_sales = np.min(sales_data[: , 1:] , axis = 1)
print(min_sales)

#To get across all restaurents per the year
min_sales = np.min(sales_data[: , 1:] , axis = 0)
print(min_sales)

#Max sales per year
min_sales = np.max(sales_data[: , 1:] , axis = 0)
print(min_sales)

#Average sales per the restaurent 
avg_sales = np.mean(sales_data[:,1:] , axis = 1)   #average also works
print(avg_sales)

#cumulative Sales column wise right  >>>>  returns 2d array right
cummsum = np.cumsum(sales_data[: , 1:] , axis = 0)
print(cummsum)

# cummulative sales row wise >>>>  returns 2d array
cummsum = np.cumsum(sales_data[: , 1:] , axis = 1)
print(cummsum)
```

```python
#Math plot 
plt.figure(figsize=(10,6))
plt.plot(np.mean(cummsum , axis=0))
plt.title("Average Cummulative Sales")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

## VECTOR OPERATIONS

```python
vector1 = np.array([1,2,3,4,5])
vector2 = np.array([6,7,8,9,10])
print(vector1 + vector2)
print(vector1 * vector2)
print(np.dot(vector1 , vector2))
angle = np.arccos(np.dot(vector1 , vector2) / ((np.linalg.norm(vector1)) * (np.linalg.norm(vector2))))
print(angle)
```
