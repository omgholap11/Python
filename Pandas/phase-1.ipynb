```python
# Merging the dataframe >> 


df1 = pd.DataFrame({
    "name" : ["om" , "arrav" , "shivam" , "anu" , "prarhana"],
     "marks" : [12,100,50,200,256]
})
df1
```

```python
#Merging datframes woth new columns like joins in sql 
df1 = pd.DataFrame({
    "name" : ["om" , "arrav" , "anu"],
     "marks" : [12,100 , 56]
})

df2 = pd.DataFrame({
    "name" : ["om" , "arrav" , "anu"],
    "roll" : [24,69 , 32]
})

#it actually applies the innner join internally columns for which no matcg are their they are skipped dircectly 
```

```python
pd.merge(df1 , df2 , on = "name")
```
